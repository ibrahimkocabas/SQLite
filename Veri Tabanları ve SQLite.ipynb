### Like ile Seçim İşlemleri

* 'a%' : “a” ile başlayan değerler
* '%a' : “a” ile biten değerler
* %or%' : herhangi bir yerinde “or” olan değerler
* '_r%' : ikinci sirasında “r” olan değerler
* 'a%%' : “a” ile başlayıp en az 3 karakterden oluşan değerler
* 'a%o' : “a” ile başlayıp “o” ile biten değerler

```python
%%sql
SELECT * FROM Customers LIMIT 10;
```

```python
%%sql
SELECT * FROM Customers WHERE CustomerName LIKE '%or%' LIMIT 10;
```

```python
%%sql
SELECT * FROM Customers WHERE CustomerName LIKE '%a' LIMIT 10;
```

```python
%%sql
SELECT * FROM Customers WHERE CustomerName LIKE 'A%' AND Country LIKE '%o';
```

## Aggregation Fonksiyonlar : Temel İstatistiksel İşlemler -- Min,Max,Sum,Avg

```python
%%sql
SELECT * FROM Products LIMIT 10;
```

```python
# Minimum ve Maksimum Değer
```

```python
%%sql
SELECT MIN(Price) AS minimum_fiyat, MAX(Price) AS maksimum_fiyat FROM Products;
```

```python
# Ortalama Değer
```

```python
%%sql
SELECT AVG(Price) AS ortalama_fiyat FROM Products
```

```python
# Toplam Fiyat
```

```python
%%sql
SELECT SUM(Price) AS toplam_fiyat FROM Products
```

```python
# Hepsini bir arada kullanma - Betimsel İstatistik Değerleri
```

```python
%%sql
SELECT COUNT(*), MIN(Price), MAX(Price), AVG(Price), SUM(Price) FROM Products;
```

## JOIN İşlemleri

Veri tabanındaki tabloları birbirine bağlamak adına işlem yapmak istediğimizde kullanılır.

```python
%%sql
SELECT name FROM sqlite_master WHERE type = 'table'
```

```python
%%sql
SELECT * FROM Categories LIMIT 3;
```

```python
%%sql
SELECT * FROM Products LIMIT 3;
```

```python
# Left Join --- SQLite için LEFT OUTER JOIN
```

Soldaki tablo ile kesişen sağdaki tablo değerlerinin bir araya getirilmesi şeklinde çalışır.

**Product tablosu ile Categories tablosunu birleştirme**

```python
%%sql
SELECT * FROM Products LEFT OUTER JOIN Categories ON Products.CategoryID = Categories.CategoryID LIMIT 5;
```

```python
# İstediğimiz Alanları Seçerek Birleştirme
```

```python
%%sql
SELECT Products.ProductName,Categories.CategoryID,Categories.CategoryName FROM Products LEFT OUTER JOIN Categories ON Products.CategoryID = Categories.CategoryID LIMIT 5;
```

```python
# Inner Join
```

İki tablo üzerinde kesisen değerleri almak istenildiği zamanlarda kullanılır.

```python
%%sql
SELECT * FROM Customers LIMIT 3;
```

```python
%%sql
SELECT * FROM Orders LIMIT 3;
```

**Sipariş veren müşterilerin SiparişID ve İsimlerine erişmek isteyelim**

```python
%%sql
SELECT Orders.OrderID, Customers.CustomerName FROM Orders INNER JOIN Customers ON Orders.CustomerID = Customers.CustomerID
```

```python
# Sipariş Sayısı
```

```python
%%sql
SELECT COUNT(*) FROM (SELECT Orders.OrderID, Customers.CustomerName FROM Orders INNER JOIN Customers ON Orders.CustomerID = Customers.CustomerID)
```

Yukarıdaki sayıdan anlaşılcağı üzere bir müşteri birden fazla sayıda sipariş vermiştir.
